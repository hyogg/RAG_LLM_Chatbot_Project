## DPR Dataset Preprocessing

### QA-pair Dataset Format (haystack)
    {
        "dataset": str,
        "question": str,
        "answers": list of str
        "positive_ctxs": list of dictionaries of format {'title': str, 'text': str, 'score': int, 'title_score': int, 'passage_id': str}
        "negative_ctxs": list of dictionaries of format {'title': str, 'text': str, 'score': int, 'title_score': int, 'passage_id': str}
        "hard_negative_ctxs": list of dictionaries of format {'title': str, 'text': str, 'score': int, 'title_score': int, 'passage_id': str}
    }


```python
import pandas as pd
import re
import json
df = pd.read_csv('./files/큐레이터 추천 소장품(25-31).csv')
df = df[['question', 'title', 'content']]
```

```python
df['answers'] = df['question'].str.split('답변: ').str[1]
df['question'] = df['question'].str.split('답변: ').str[0]
df
```

```python
# 필요없는 문자열 삭제
for idx, row in df.iterrows():
    df.at[idx, 'question'] = row['question'].replace('질문: ', '').replace('\n', '')
    df.at[idx, 'answers'] = row['answers'].replace('질문: ', '')
    df.at[idx, 'content'] = re.sub(r'page_content=|\[|\]|\"|\'|\,', '', row['content'])
df
```

```python
df = df.rename(columns={'content':'text'})
df = df[['question', 'answers','title', 'text']]
df
```

```python
sdf = df.copy()
```

```python
from rank_bm25 import BM25Okapi

json_data = []
corpus = sdf['text']

tokenized_corpus = [doc.split(" ") for doc in corpus]
bm25 = BM25Okapi(tokenized_corpus)
for index, row in sdf.iterrows() :
    query = row['question']
    tokenized_query = query.split(" ")
    doc_scores = bm25.get_scores(tokenized_query)
    score = doc_scores[index]
     
    # 나중에 negative 넣을 때 쓸 코드
    # if query == row['question'] :
    #     answers = row['answers']
    #     pos_title = row['title']
    #     pos_score = row['bm25_score']
    # else
        # neg_title = row['title']
        # neg_score = row['bm25_score']
    # sorted_doc_scores = sorted(doc_scores, reverse=True)
    # sorted_df = sdf.sort_values(by='bm25_score', ascending=False)
    # for i, (array, series) in enumerate(zip(sorted_doc_scores, sorted_df.bm25_score.values)) :
    #     assert np.array_equal(array, series), f'다름: {i+1}행 참고'
    # print('모든 행이 일치함')

    json_row = {
        "dataset": 'split_25p',
        "question": query,
        "answers": [row['answers']],
        "positive_ctxs": [
            {
                'title': row['title'],
                'text': row['text'],
                'score': score,
                'title_score': 0, 
                'passage_id': ''
            }
        ],
        "negative_ctxs": [],
        "hard_negative_ctxs": []
    }
    json_data.append(json_row)
    print(index, '번째 row 출력')
json_data[0]
```
