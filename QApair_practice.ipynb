```python
import pandas as pd
import re
df = pd.read_csv('./files/큐레이터 추천 소장품(25-31).csv')
df
```

```python
df['answers'] = df['question'].str.split('답변: ').str[1]
df['question'] = df['question'].str.split('답변: ').str[0]
df
```

```python
for idx, row in df.iterrows():
    df.at[idx, 'question'] = row['question'].replace('질문: ', '').replace('\n', '')
    df.at[idx, 'answers'] = row['answers'].replace('질문: ', '')
    df.at[idx, 'content'] = re.sub(r'page_content=|\[|\]|\"|\'|\,', '', row['content'])
df
```

```python
df[df['question'].str.contains('다\.', regex=True) | df['answers'].str.contains('\?', regex=True)]
```

```python
df = df.rename(columns={'content':'text'})
df = df[['question', 'answers','title', 'text']]
df
```

```python
from rank_bm25 import BM25Okapi

corpus = [
    "Hello there good man!",
    "It is quite windy in London",
    "How is the weather today?"
]

tokenized_corpus = [doc.split(" ") for doc in corpus]
bm25 = BM25Okapi(tokenized_corpus)
```

```python
query = "windy London"
tokenized_query = query.split(" ")

doc_scores = bm25.get_scores(tokenized_query)
doc_scores
# array([0.        , 0.93729472, 0.        ])
```
